**Collecting the Details of Timeline Tweets in a Dataframe**

```python
def  timeline_df(tweets):
    import pandas as pd
    df = pd.DataFrame()
    df['id'] = list(map(lambda tweet: tweet.id, tweets))
    df['created_at'] = list(map(lambda tweet: tweet.created_at, tweets))
    df['text'] = list(map(lambda tweet: tweet.text, tweets))
    df['favorite_count'] = list(map(lambda tweet: tweet.favorite_count, tweets))
    df['retweet_count'] = list(map(lambda tweet: tweet.retweet_count, tweets))
    df['source'] = list(map(lambda tweet: tweet.source, tweets))
    return df
```

```python
df = timeline_df(tweets)
df.head()
```

```python
def group(prop):
    import pandas as pd
    ids = [i.id for i in prop]
    df = pd.DataFrame(ids, columns = ["id"])
    df["created_at"] = [i.created_at for i in prop]
    df["screen_name"] = [i.screen_name for i in prop]
    df["location"] = [i.location for i in prop]
    df["followers_count"] = [i.followers_count for i in prop]
    df["statuses_count"] = [i.statuses_count for i in prop]
    df["friends_count"] = [i.friends_count for i in prop]
    df["favourites_count"] = [i.favourites_count for i in prop]
    return df
```

**Feature Engineering for Segmentation of the Followers**

For developing targeted communication strategies, we want to segmentate the followers based on a score that accounts both their activity and impact on the tweeter, i.e., a feature proportional to the number of likes and their tweet number.   

```python
df_fol.index = df_fol["screen_name"]
df_fol.head()
```

```python
df_features = df_fol[["followers_count", "statuses_count"]]
df_features.head()
```

As the data on these features are quite diverse, let us normalize them.

```python
df_features = df_features.apply(lambda x: (x-min(x))/(max(x)-min(x)))
```

When these two features are multiplied, if one feature is zero, the effect of the other feature disappears fully. This can be avoided by translating all the counts by a certain amount.

```python
df_features["followers_count"] = df_features["followers_count"] + 0.01
df_features["statuses_count"] = df_features["statuses_count"] + 0.01
```

Assuming both features have equal weight/importance, let us develop a score from their product. 

```python
score = df_features["followers_count"] * df_features["statuses_count"]
```

```python
score
```

Let us add score in the dataframe.

```python
df_features["score"] = score
df_features.head()
```

Let us call scores larger than the sum of median and std. deviation per the number of followers.

```python
score[score > score.median() + score.std()/len(score)].sort_values(ascending = False)
```

Let us label this class of most active users as A, and the others as B.

```python
df_features["segment"] = np.where(df_features["score"] >= score.median() + score.std()/len(score), "A","B")
```
